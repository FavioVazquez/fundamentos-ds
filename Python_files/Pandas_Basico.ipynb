## Joins

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                  'A': ['A0', 'A1', 'A2', 'A3'],
                  'B': ['B0', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']})

# Inner by default
pd.merge(left, right, on='key')
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                  'key2': ['K0', 'K1', 'K0', 'K1'],
                  'A': ['A0', 'A1', 'A2', 'A3'],
                  'B': ['B0', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                   'key2': ['K0', 'K0', 'K0', 'K0'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']})


pd.merge(left, right, on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='inner', on=['key1', 'key2'])
```

## Transformaciones avanzadas e ingeniería de características

La información de los datos es la siguiente:

- datetime: hora por hora + marca de tiempo

- season:
    - 1 = primavera
    - 2 = verano
    - 3 = caída
    - 4 = invierno

- holiday: si el día se considera un día festivo

- workingday: si el día no es un fin de semana ni un feriado.

- clima:

    - 1: Despejado, pocas nubes, medio nublado, parcialmente nublado
    - 2: Niebla + Nublado, Niebla + parcialmente nublado, Niebla + Pocas nubes, Niebla
    - 3: Nieve ligera, Lluvia ligera + Tormenta eléctrica + Nubes dispersas, Lluvia ligera + Nubes dispersas
    - 4: Lluvia intensa + Paletas de hielo + Tormenta + Niebla, Nieve + Niebla


- temp: temperatura en grados Celsius

- atemp: "sensación térmica" la temperatura en grados Celsius

- humidity: humedad relativa

- windspeed: velocidad del viento

- casual: número de alquileres de usuarios no registrados iniciados

- registered: número de alquileres de usuarios registrados iniciados

- count: número de alquileres totales

### Tyding Pandas dataframes

- Cada variable en el conjunto de datos se coloca en su propia columna.
- Cada observación se coloca en su propia fila.
- Cada valor se coloca en su propia celda.

## Trabajando con fechas

Trabajar bien con fechas en dataframes es crucial en el trabajo de un científico de datos. Entendamos cómo funciona:

```python
df = pd.read_csv("data/bikes.csv")
```

```python
import calendar
from datetime import datetime
import numpy as np

df["date"] = df.datetime.apply(lambda x : x.split()[0])
```

```python
help(df.apply)
```

```python
df.datetime.apply(lambda x : x.split()[0])[1], df.datetime[1]
```

```python
df["hour"] = df.datetime.apply(lambda x : x.split()[1].split(":")[0])
```

```python
df["weekday"] = df.date.apply(lambda date_string : calendar.day_name[datetime.strptime(date_string,"%Y-%m-%d").weekday()])
```

```python
df["month"] = df.date.apply(lambda date_string : calendar.month_name[datetime.strptime(date_string,"%Y-%m-%d").month])
```

```python
df["season"] = df.season.map({1: "Spring", 2 : "Summer", 3 : "Fall", 4 :"Winter" })
```

```python
df["weather"] = df.weather.map({1: " Clear + Few clouds + Partly cloudy + Partly cloudy",\
                                        2 : " Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist ", \
                                        3 : " Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds", \
                                        4 :" Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog " })
```
